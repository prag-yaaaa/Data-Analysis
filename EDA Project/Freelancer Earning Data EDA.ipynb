The Following data is taken from KAGGLE.

The data taken is related to tracking of FREELANCER EARNINGS and the JOB TRENDS that they have in the market across multiple platforms, experience levels and job categories. It provides valuable insights into compensation patterns, client preferences, and success metrics in the global freelance marketplace, helping freelancers, businesses and researchers understand the dynamics of the gig economy. 


Importing Packages

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

DATA LOADING

Reading the dataset using pandas library into the dataframe

```python
df = pd.read_csv("freelancer_earnings_bd.csv")
```

You can see the columns that the imported file has and the example data values that each column has. Sample rows of the dataset is shown. Here, 

Freelancer_ID is a unique identifier for each freelancer record.

Job_Category is Promary classification of freelance work performed.

Platform is Freelance marketplace where work was performed.

Experience_Level is Freelancer's professional experience tier.

Client_Region is Geographical Location of the client.

Payment_Method is method used for financial transactions.

Job_Completed is number of projects successfully completed.

Earnings_USD is the total earnings in US Dollars.

Hourly_Rate represents Freelancer's hourly compensation rate in USD.

Job_Success_Rate is percentage of successful Job completions.

Client_Rating is average rating given by clients (1.0-5.0 scale)

Job_Duration_Day is average project timeline in days.

Project_Type is the classification of work arrangement.

Rehire_Rate is the percentage of clients who rehire the freelancers.

Marketing_Spend is the amount invested in platform promotion in USD.

OUTLIER Detection Using Boxplot

```python
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']
print('We have {} numerical features : {}'.format(len(numeric_features),numeric_features))
print('\nWe have {} categorical features: {}'.format(len(categorical_features), categorical_features))

plt.figure(figsize = (20,20))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize = 20, fontweight = 'bold', alpha = 0.8, y = 1.)
df[numeric_features].plot(kind = 'box', subplots = True, layout = (4,5), figsize = (20,20))

plt.show()
```

No outliers detected.

```python
grouped_data = df.groupby(["Experience_Level", "Job_Category"])["Earnings_USD"].mean().reset_index()
plt.figure(figsize = (10,5))
custom_palette = ["#073763","#34516a","#3f6888","#7896b2","#aedde6","#696969","#122936","#1d4257"]
sns.barplot(data=grouped_data, x="Experience_Level", y="Earnings_USD", hue="Job_Category", palette= custom_palette)
plt.legend(title ='Job Category', loc='center left', bbox_to_anchor=(1, 0.5))
plt.title("Average Earnings by each Experience level distribution with job category")
plt.show()
```

Findings:

2. Beginners (0) have competitice earnings in creative and marketing fields, suggesting that experience isnt always the primary factor.
3. Web & App Development are the highest paying categories, reinforcing the importance of technical skills.
4. SEO, Data Entry and Content Writing have lower earnings due to high competition and price sensitivity in the market.

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title ("Relationship Between Numerical Features", pad = 20)
plt.show()
```

Findings :
1. No highly correlated variables - Independent features with no direct influence on each other. Most of the correlation values are close to 0, indicating very weak or no linear relationship between numerical variables.
2. Weak relationships suggest non-linear or external factors impacting earnings.
3. Further analysis (like regression, clustering) may be needed to find deeper patterns.
4. Slightly positive correlations is seen between : Experience_Level and Earnings_USD, Hourly_Rate and Earnings_USD, Marketing_Spend and Earnings_USD, JOB_Completed and Hourly_Rate.
5. Since no strong correlations exist, earnings are not strongly driven by a single numerical factor.
6. Hourly rate VS. Earnings has a slight correlation, meaning higher rates do not gurantee higher earnings - perhaps due to fewer job opportunities at higher proces.
7. Experience level has a minor impact on earnings, suggesting that other factors like skills, job success rate or client relationships might be more crtical.
